```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images')
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        ret, frame = cap.read()
        
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('Image Collection', frame)
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

with open("drowsiness_log.txt", "w") as f:
    f.write("Drowsiness Detection Log\n")
    f.write("Timestamp\n")

def log_drowsiness_detection():
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open("drowsiness_log.txt", "a") as f:
        f.write(timestamp + "\n")

def draw_alert(frame):
    alert = "Drowsiness Detected! Take a break."
    cv2.putText(frame, alert, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    
while cap.isOpened():
    ret, frame = cap.read()
    
    results = model(frame)

    cv2.imshow('Drowsiness Detection', np.squeeze(results.render()))
        
    for detection in results.pred[0]:
        if detection['label'] == 'drowsy':
            draw_alert(frame)
            log_drowsiness_detection()
            break

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
